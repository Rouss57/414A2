```python
import json

# Replace 'your_file.json' with the name of your JSON file
file_name = 'test.json'

# Open and load the JSON data
with open(file_name, 'r') as file:
    data = json.load(file)


```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a new graph
G = nx.Graph()

# Limit the number of recipes to process
recipe_limit = 25

# Add nodes and edges from the data
for item in data[:recipe_limit]:
    recipe_id = item['id']
    ingredients = item['ingredients']
    
    # Add the recipe node
    G.add_node(recipe_id, type='recipe')
    
    # Add ingredient nodes and edges between recipe and ingredients
    for ingredient in ingredients:
        G.add_node(ingredient, type='ingredient')
        G.add_edge(recipe_id, ingredient)

# Visualize the graph
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw_networkx_nodes(G, pos, node_size=700, cmap=plt.get_cmap('jet'), 
                       node_color=[G.degree(v) for v in G])
nx.draw_networkx_labels(G, pos, font_size=12)
nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.title("Web Graph of Recipes and Ingredients (Subset)")
plt.show()
```
